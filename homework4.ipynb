W4A8 :ReverseAddPalindrome]
Nhập số nguyên dương n. Lặp lại phép toán n = n + reverse(n) cho đến khi n là số đối xứng (palindrome). In ra số bước và giá trị palindrome thu được.

```python
n = int(input())
buoc = 0   #chuỗi sô chưa bằng số đảo ngược thì tiếp tục lặp
while str(n)!= str(n)[::-1]: #biến số thành chuỗi số rồi đảo ngược 
    reverse = int(str(n)[::-1]) #reverse ; đảo xong chuyển lại thành số 
    n = n + reverse
    buoc += 1  #đếm số lần thực hiện phép cộng
print("Số bước:",buoc)
print("Số đối xứng:",n)





```

W4A10 :"[LongestCollatzUpToN]
Nhập vào số nguyên dương n. Tìm số x trong [1, n] có độ dài dãy Collatz lớn nhất và in ra x cùng độ dài dãy tương ứng (x L). Nếu có nhiều số cùng độ dài, chọn số
nhỏ nhất."

```python
n = int(input())
max_len = 0 #độ dài dãy collatz lớn nhất tìm đc 
best_x = 1 #Số x tương ứng với max_len
for x in range(1,n+1):  #xét 1 số x 
    temp = x 
    length = 1 
    while temp!=1:
        if temp%2==0:
            temp//=2
        else:
            temp = 3*temp + 1
        length += 1 
    if length > max_len :
        max_len = length 
        best_x = x
print(best_x, max_len)


```

Số này bằng tổng tất cả các ước của số kia (trừ chính nó) và ngược lại.
Ví dụ 220 và 284 là một cặp số thân thiết
220 ngoài chính nó ra, nó còn có 11 ước số khác là: 1, 2, 4, 5, 10, 11, 20, 44, 55,
110. Tổng của 11 ước này cho chúng ta 284
284 ngoài bản thân nó, còn có 5 ước số khác là: 1, 2, 4, 71, 142. Tổng của chúng
vừa đúng bằng 220
Nhập hai số từ bàn phím và kiểm tra xem chúng có phải là cặp số thân thiết không.
Nếu có, in ra màn hình "true", ngược lại, in ra "false" 

```python
a, b = map(int,input().split())
tong_a = 0 
for i in range(1,a):
    if a%i==0 :
        tong_a += i
tong_b = 0 
for k in range(1,b):
    if b%k==0:
        tong_b += k
if tong_a == b and tong_b == a:
    print("true")
else :
    print("false")
```
