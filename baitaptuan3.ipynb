W3A1 : Nhập vào một số nguyên dương có 3 chữ số. Tìm và in ra màn hình “SỐ
NGUYÊN” đảo ngược của n. Không dùng kiểu chuỗi. Chú ý: kiểm tra chỉ xử
lý các số nguyên dương có đúng 3 chữ số. Các số khác không xử lý.

```python
n = int(input())
if n>0 and 100<= n <=999 :
    hang_don_vi = n%10
    hang_chuc = (n//10)%10
    hang_tram = n//100 
    so_dao_nguoc = hang_don_vi*100 + hang_chuc*10 + hang_tram
    print(f"Số đảo ngược của số đã nhập là :{so_dao_nguoc}")
else :
    print("Eror.Sai định dạng!") 
```

W3A3 :Kiểm tra xem một số có phải là lũy thừa của 2 hay không
Sử dụng các toán tử bitwise để xác định xem một số có phải là lũy thừa của 2
hay không.
Gợi ý: Đối với lũy thừa của 2, n & (n-1) == 0

```python
def la_luy_thua_2(n):
    return n>0 and (n&(n-1)) == 0 
n = int(input())
print(la_luy_thua_2(n))
```

W3A14 : [Giải phương trình bậc nhất]

nghiệm của phương trình, kết quả làm tròn đến số thập phân thứ 2. In ra "Vô
nghiệm" hoặc "Vô số nghiệm" trong trường hợp tương ứng

```python
a, b = map(float,input().split())
if a!= 0 :
    if b == 0 :
        print("Phương trình có nhiệm duy nhất là : x = 0 ")
    else : 
        x = -b/a 
        print(f"Phương trình có nhiệm duy nhất : x = {x:.2f}")
else :
    print("Phương trình vô nhiệm ")
```

W3A15 : [Xếp loại học lực]
Nhập vào điểm trung bình của một học sinh và in ra học lực của học sinh đó.
Xếp loại học lực theo quy tắc:
- >= 8.0: Giỏi
- >= 6.5: Khá
- >= 5.0: Trung bình
- < 5.0: Yếu

```python
n = float(input())
if n > 0 : 
    if n>=8.0 :
        print("Giỏi")
    elif 6.5<=n<8.0 :
        print("Khá")
    elif 5.0 <= n <6.6 :
        print("Trung bình ")
    elif n<5.0 :
        print("Yếu")
else :
    print("Eror.Nhập sai định dạng")
```
