W7A1 : Viết hàm tìm kiếm nhị phân trên một danh sách đã được sắp xếp tăng dần. Trả về chỉ số của phần tử hoặc -1 nếu không tìm thấy. mẫu hàm: def binary_search(arr, left, right, target):


```python
def binary_search(arr, left, right, target):     #tìm kiếm nhị phân là cách tìm một phần tử 
    while left <= right:                          # rất nhanh trong danh sách đc sắp xếp
        mid = (left + right)//2 
        if arr[mid] == target :
            return mid 
        elif arr[mid] < target:
            left = mid + 1 
        else :
            right = mid - 1
    return -1 
arr = list(map(int,input().split()))
target = int(input())
print(binary_search(arr, 0 , len(arr) -1 , target))
```

W7A3 : Viết hàm thực hiện thuật toán bubble sort, Selection Sort để sắp xếp danh sách theo thứ tự tăng dần. Sắp xếp nổi bọt (Bubble Sort); Sắp xếp chọn (Selection Sort)


```python
def bubble_selecction_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(n-1-i):
            if  arr[j] > arr[j+1] :
                arr[j], arr[j+1] = arr[j+1], arr[j] #không đc dùng return trong vòng lặp, hoán đổi 
    return arr                                      # không phải trả về giá trị 
arr = list(map(int,input().split()))
print(bubble_selecction_sort(arr))
              
```

W7A6 : Bạn được cung cấp một danh sách các danh sách lồng nhau, vd [[1, 2, 3], [4,5], [6, 7, 8, 9]], hãy làm phẳng danh sách               flatten : làm phẳng


```python
def flatten(lst) :
    result = []    #danh sách rỗng chứa phần tử sau khi là phẳng
    for x in lst:  # duyệt phần tử trng lst
        if type(x) == list:
            temp = flatten(x)
            for x in temp : 
                result.append(x)    #là phẳng
        else : 
            result.append(x)
    return result 
lst = [[1,2,3], [4,5], [6,7,8,9]]
print(flatten(lst))


```

```python
nums = list(map(int,input().split()))
n = len(nums)
dp = [1] * n
for i in range(n) :
    for j in range(i) :
        if nums[i] > nums[j] :
            dp[i] > max(dp[i],dp[i]+1)
print(max(dp))
```

W7A8 : Bạn được cung cấp: một danh sách các khoảng thời gian interval = [[l1, r1], [l2, r2], ...] và một danh sách các truy vấn queries = [q1, q2, ...] Với mỗi truy vấn q, hãy tìm độ dài nhỏ nhất của khoảng thời gian [li, ri] sao cho li ≤ q ≤ ri. Nếu không tồn tại khoảng thời gian nào như vậy, hãy trả về -1 cho truy vấn đó. solve in < O(m.n), hint: binary search in the queries list


```python
import bisect   #hàm tìm kiếm nhị phân
def giai_bai_toan(cac_khoang, cac_truy_van) :
    so_luong_truy_van = len(cac_truy_van) # 1.khởi tạo mảng kết quả ban đầu toàn là -1
    ket_qua = [-1] * so_luong_truy_van # 2. Sắp xếp các khoảng thời gian theo độ Dài tăng dần (r-1), khoảng nào ngắn hơn đc ưu tiên xử lý trc
    cac_khoang.sort(key=lambda x: x[1] - x[0])
    truy_van_sap_xep = sorted([(q, i) for i, q in enumerate (cac_truy_van)])  #đính kèm vị trí(chỉ số) gốc vào từng truy vẫn rồi tăng dần
    mang_gia_tri = [phan_tu[0] for phan_tu in truy_van_sap_xep]
    for l,r in cac_khoang :
        do_dai = r - l
        vi_tri_dau = bisect.bisect_left(mang_gia_tri, l) #tìm vtri đầu tiên >=l
        vi_tri_cuoi = bisect.bisect_right(mang_gia_tri, r) #tìm ví trị đầu >r
        for i in range(vi_tri_dau, vi_tri_cuoi) :
            vi_tri_goc = truy_van_sap_xep[i][1]
            if ket_qua[vi_tri_goc] == -1 :
                ket_qua[vi_tri_goc] = do_dai
    return ket_qua 
cac_khoang = [[1,4], [2,4], [3,6]]
cac_truy_van = [2, 3, 5]
In_ket_qua = giai_bai_toan(cac_khoang, cac_truy_van)
print("Kết quả trả về:", In_ket_qua)




```

```python
import bisect 
nums = list(map(int,input().split()))   # 5 2 6 1 
n = len(nums)
ket_qua = [0]*n   #khởi tạo  mảng kết quả toàn số 0 
mang_da_duyet = []  #lưu số các số bên phải đã duyệt & sắp xếp tăng dần
for i in range(n-1, -1, -1): #duyệt ngc từ cuối về đầu  1-6-2-5
    so_hien_tai = nums[i]
    vi_tri = bisect.bisect_left(mang_da_duyet, so_hien_tai) #tìm kiếm nhị phân xem số hiện tại lên đứng ở vtri nào, vị trí là số lượng phần tử nhỏ hơn nó ở bên phải 
    ket_qua[i] = vi_tri   #lưu số tìm đc vào kết quả 
    bisect.insort(mang_da_duyet, so_hien_tai)  #chèn số hiện tại vào mảng đã duyệt , giữ nguyên thứ tự sắp xếp 
print("Kết quả số lượng phải nhỏ hơn:", ket_qua)

```

```python
from bisect import bisect_left , insort     # bisert_left(tìm vị trí), insort(chèn giữ nguyên thứ tự)
nums = list(map(int,input().split()))
sorted_list = []     # mảng phụ chứ số bên phải đã duyệt qua , tự dộng sắp xếp tăng dần
result = []          # mảng trống gom kết quả đếm qua từng bước
for x in reversed(nums):   #reversed(Nums) : danh sách đảo ngược của nums 1-6-2-5
    count = bisect_left(sorted_list,x)  # tìm kiếm nhị phân để xem số x hiện tại đạt trong sorted_list(đang tăng dần) thì sẽ đứng ở vị tr chỉ số (index) bao nhieu 
    result.append(count)                # mảng tăng dần thì đúng bằng số lượng phần tử nhỏ hơn x trong sorted_list 
    insort(sorted_list,x)  # chèn x vừa tìm đc vào sorted_list
result.reverse()    # trả về đúng thứ tự
print(result)
```

W7A10 : Cho một danh sách các số nguyên nums và một số nguyên m, hãy tìm tổng mảng con lớn nhất theo modulo m. Tức là, với bất kỳ mảng
con nums[i:j] nào, hãy tính sum(nums[i:j]) % m và trả về kết quả lớn nhất có thể. Gợi ý: Duy trì tổng tiền tố (prefix sum) mod m và sử dụng tìm kiếm nhị phân để tìm tổng tiền tố nhỏ nhất chỉ lớn hơn tổng hiện tại một chút (để tối đa hóa hiệu số bao quanh).
>tìm tổng mảng con lớn nhất sau khi lấy dư cho m 


```python
from bisect import bisect_right, insort
nums = list(map(int,input().split()))
m = int(input())
prefix = 0 
best = 0 
sorted_prefix = []
for x in nums : 
    prefix = (prefix + x) % m 
    best = max(best, prefix)
    idx = bisect_right(sorted_prefix, prefix)
    if idx < len(sorted_prefix):
        candidate = (prefix - sorted_prefix[idx]+m ) % m
        best = max(best, candidate)
    insort(sorted_prefix, prefix)
print(best)
```

```python
arr = list(map(int,input().split()))
d = {}  #tạo từ điển để lưu dạng số:số lần xuất hiện
for x in arr :
    if x in d :
        d[x] += 1 
    else : 
        d[x] = 1 
ans = 0 
for x in d : #duyệt qua key trong d 
    if x + 1 in d :
        ans = max(ans, d[x] + d[x+1])
print(ans)
```

W7A16 : Đề bài
Cho một mảng số nguyên không âm. Gọi mảng đó là "mảng đẹp" nếu
mỗi phần tử x trong mảng xuất hiện đúng x lần. Kiểm tra xem mảng có phải là mảng đẹp hay không. Nếu không hãy xóa đi y số (y là nhỏ nhất có thể) để mảng còn lại thành mảng đẹp. 
> đếm số lần xuất hiện từng số ; với mỗi số x: - x lần -> giữ nguyên
>                                              - ít hơn x lần   -> xóa hết 
>                                              - nhiều hơn x lần -> xóa bớt còn đúng x lần
                                            


```python
arr = list(map(int,input().split()))  #1,2,2,3,3,3,3
d = {}
for x in arr : 
    d[x] = d.get(x,0) + 1 # đếm số lần xuất hiện
ans = 0    #tích lũy số phần tử phải xóa
for x, cnt in d.items():     #cnt:count_số lần xuất hiện của số đó 
    if cnt < x : #xuất hiện ít hơn x lần > xóa hết
        ans += cnt   # tất cả lần xuất hiện bị xóa đi
    elif cnt > x : #xuất hiện nhiều hơn x lần > xóa bớt 
        ans += cnt - x # xóa đi phần thừa
print("Tổng số phần tử cần xóa là ",ans)
```

>Thuật toán này hoạt động nhưu 1 cuộc bầu cử -candidate:lưu trữ ứng viên hiện tại đang dẫn đầu cuộc bầu cử 
>                                            -count:lưu số phiếu vượt trội ứng viên đó so với các số khác 

```python
arr = list(map(int,input().split()))
candidate = None  # Ban đầu chưa có ai ứng cử 
count = 0         # số phiêu bấu = 0
for x in arr:      # vòng lặp Boyer-Moonre(tìm ứng viên sáng giá )
    if count == 0 :    #số phiếu phe cũ bằng 0 
        candidate = x  # bầu x là ứng viên 
        count = 1      # cấp cho x phiếu bầu đầu tiên
    elif x == candidate :   # số x đang xét trùng với x hiện tại _được bầu 
        count += 1          # thêm 1 phiếu ủng hộ
    else:                   # số đang xét khác , mất 1 phiêu đổi kháng
        count -= 1          # trừ 1 phiếu 
if arr.count(candidate) > len(arr)//2:   #arr.count(candidate):xem xuất hiện bao nhiêu lần , lớn hơn 1/2 thắng cử 
    print(candidate)
# input[2,2,1,1,2]
```

```python
a = list(map(int,input().split()))
b = list(map(int,input().split()))
result = []
i = 0 
j = 0 
tang = True 
if len(a) > 2 and a[0] > a[1]:
    tang = False
while i < len(a) and j < len(b):
    if tang: #tăng dần
        if a[i] <= b[j]:
            result.append(a[i])
            i += 1 
        else :
            result.append(b[j])
            j += 1 
    else : #giảm dần 
        if a[i] <= b[j] :
            result.append(a[i])
            i += 1 
        else:
            result.append(b[j])
            j += 1 
while i < len(a):   #phần còn lại
    result.append(a[i])
    i += 1 
while j < len(b):
    result.append(b[j])
    j += 1 
print(*result)
    
```

Cho N tên sinh viên gồm cả họ và tên, tìm xem tên nào hay gặp nhất.
Nếu có nhiều tên cùng số lần gặp thì lấy  có thứ tự từ điển nhỏ nhất

```python
from collections import Counter 
names = [name.strip() for name in input().split(",") ]
freg = Counter(names)                                                             #đếm số lần xuất hiện từng tên 
max_count = max(freg.values())                                                    #Tìm số lần cuất hiện nhiều nhất 
most_frequent_names = [name for name , count in freg.items()if count == max_count]#tên có số lần xuất hiện = max_count
best_name = min(most_frequent_names) #tên có thứ tự từ điển min
print(best_name)

```

Tìm số có đúng k chữ số lớn nhất có thể trong một số có n chữ số (coi đầu vào là một số rất lớn - str). Không được hoán đổi vị trí các chữ số.


```python
s = input()
k = int(input())
stack = []
remove = len(s) - k 
for c in s : 
    while stack and remove > 0 and stack[-1] < c :
        stack.pop()
        remove -= -1 
    stack.append(c)
print(''.join(stack[:k]))
```

Bà̀i toán tìm dãy các hợp số dài nhất liên tiếp không chứa số nguyên nào <= N (N là số nguyên dương cho trước). Nếu có nhiều dãy cùng độ dài thì lấy dãy nhỏ nhất (gặp lần đầu tiên từ trái qua)
>tìm và in ra dãy các hợp số liên tiếp dài nhất nằm trong đoạn N+1->2N
(N nhập từ bàn phím)_hợp số là số tự nhiên lớn hơn 1 và có nhiều hơn 2 ước (ngoài 1 & chính nó) , số 0,1,số nguyên tố ko phải hợp số

```python
def hop_so(x):
    if x < 4 :  #(0,1,2,3)
        return False 
    for i in range(2, int((x)**0.5)+1):  #số nguyên tố 
        if x % i == 0 :    #ko phải số nguyên tố thì là hợp số
            return True 
    return False 
N = int(input())
max_len = 0 #Độ dài dãy hợp số liên tiếp dài nhất tìm thấy từ trc đến nay
cur_len = 0  #Độ dài dãy hợp số hiện tại đang đếm 
start = 0  #số đầu tiên của dãy hợp số dài nhất
cur_start = 0  #
for x in range(N + 1, 2 * N +1):
    if hop_so(x):
        if cur_len == 0 :
            cur_start = x 
        cur_len  += 1 
        if cur_len > max_len :
            max_len = cur_len 
            start = cur_start 
    else : 
        cur_len = 0 
for x in range(start, start + max_len):
    print(x, end=" ")


```

Tìm Dãy con tăng dần liên tục dài nhất của một danh sách nhập vào từ bàn phím. Lấy dãy đầu tiên nếu có nhiều dãy có số phần tử bằng nhau

Tìm Dãy con tăng dần KHÔNG liên tục dài nhất của một danh sách nhập vào từ bàn phím

```python
arr = list(map(int,input().split()))
n = len(arr)
dp = [1] * n
for i in range(n):
    for j in range(i):
        if arr[i]>arr[j]:
            dp[i] = max(dp[i], dp[j]+1)
print(max(dp))

```
