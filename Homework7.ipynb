W7A1 : Viết hàm tìm kiếm nhị phân trên một danh sách đã được sắp xếp tăng dần. Trả về chỉ số của phần tử hoặc -1 nếu không tìm thấy. mẫu hàm: def binary_search(arr, left, right, target):


```python
def binary_search(arr, left, right, target):     #tìm kiếm nhị phân là cách tìm một phần tử 
    while left <= right:                          # rất nhanh trong danh sách đc sắp xếp
        mid = (left + right)//2 
        if arr[mid] == target :
            return mid 
        elif arr[mid] < target:
            left = mid + 1 
        else :
            right = mid - 1
    return -1 
arr = list(map(int,input().split()))
target = int(input())
print(binary_search(arr, 0 , len(arr) -1 , target))
```

W7A3 : Viết hàm thực hiện thuật toán bubble sort, Selection Sort để sắp xếp danh sách theo thứ tự tăng dần. Sắp xếp nổi bọt (Bubble Sort); Sắp xếp chọn (Selection Sort)


```python
def bubble_selecction_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(n-1-i):
            if  arr[j] > arr[j+1] :
                arr[j], arr[j+1] = arr[j+1], arr[j] #không đc dùng return trong vòng lặp, hoán đổi 
    return arr                                      # không phải trả về giá trị 
arr = list(map(int,input().split()))
print(bubble_selecction_sort(arr))
              
```

W7A6 : Bạn được cung cấp một danh sách các danh sách lồng nhau, vd [[1, 2, 3], [4,5], [6, 7, 8, 9]], hãy làm phẳng danh sách               flatten : làm phẳng


```python
def flatten(lst) :
    result = []    #danh sách rỗng chứa phần tử sau khi là phẳng
    for x in lst:  # duyệt phần tử trng lst
        if type(x) == list:
            temp = flatten(x)
            for x in temp : 
                result.append(x)    #là phẳng
        else : 
            result.append(x)
    return result 
lst = [[1,2,3], [4,5], [6,7,8,9]]
print(flatten(lst))


```

W7A7 : Cho một danh sách các số nguyên nums, hãy tìm độ dài của dãy con tăng nghiêm ngặt dài nhất (không nhất thiết phải liền kề)._Longest Increasing Subsequence(LIS)



```python
nums = list(map(int,input().split()))
n = len(nums)
dp = [1] * n
for i in range(n) :
    for j in range(i) :
        if nums[i] > nums[j] :
            dp[i] > max(dp[i],dp[i]+1)
print(max(dp))
```

W7A28
Bà̀i toán tìm dãy các hợp số dài nhất liên tiếp không chứa số nguyên nào <= N (N là số nguyên dương cho trước). Nếu có nhiều dãy cùng độ dài thì lấy dãy nhỏ nhất (gặp lần đầu tiên từ trái qua)
>tìm và in ra dãy các hợp số liên tiếp dài nhất nằm trong đoạn N+1->2N
(N nhập từ bàn phím)_hợp số là số tự nhiên lớn hơn 1 và có nhiều hơn 2 ước (ngoài 1 & chính nó) , số 0,1,số nguyên tố ko phải hợp số

```python
def hop_so(x):
    if x < 4 :  #(0,1,2,3)
        return False 
    for i in range(2, int((x)**0.5)+1):  #số nguyên tố 
        if x % i == 0 :    #ko phải số nguyên tố thì là hợp số
            return True 
    return False 
N = int(input())
max_len = 0 #Độ dài dãy hợp số liên tiếp dài nhất tìm thấy từ trc đến nay
cur_len = 0  #Độ dài dãy hợp số hiện tại đang đếm 
start = 0  #số đầu tiên của dãy hợp số dài nhât
cur_start = 0 
for x in range(N + 1, 2 * N +1):
    if hop_so(x):
        if cur_len == 0 :
            cur_start = x 
        cur_len  += 1 
        if cur_len > max_len :
            max_len = cur_len 
            start = cur_start 
    else : 
        cur_len = 0 
for x in range(start, start + max_len):
    print(x, end=" ")


```

W7A29
Tìm Dãy con tăng dần liên tục dài nhất của một danh sách nhập vào từ bàn phím. Lấy dãy đầu tiên nếu có nhiều dãy có số phần tử bằng nhau

```python
arr = list(map(int,input().split()))   #[1,2,3,1,2]
max_len = 1 
cur_len = 1
start = 0    # vị trí bắt đầu dãy dài nhất 
cur_start = 0  # vị trí bắt đầu dãy hiện tại 
for i in range(1, len(arr)):
    if arr[i] > arr[i-1]:
        cur_len += 1 
    else :
        cur_len = 1 
        cur_start = i    # chỉ cập nhật khi lớn hơn , nếu bằng nhau giữ dãy đầu tiên
    if cur_len > max_len:
        max_len = cur_len 
        start = cur_start
print(arr[start:start + max_len])
```

Tìm Dãy con tăng dần KHÔNG liên tục dài nhất của một danh sách nhập vào từ bàn phím

```python
arr = list(map(int,input().split()))
n = len(arr)
dp = [1] * n
for i in range(n):
    for j in range(i):
        if arr[i]>arr[j]:
            dp[i] = max(dp[i], dp[j]+1)
print(max(dp))

```
