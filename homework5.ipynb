thể ánh xạ thành chính nó. Ví dụ, từ "abca" và "zbxz" là hai từ đẳng cấu vì ta có thể ánh xạ 'a' thành
'z', 'b' thành 'b' và 'c' thành 'x'.
Viết hàm trả về \(true\) nếu hai từ \(a\) và \(b\) là đẳng cấu, ngược lại \(false\). Bạn có thể giả sử hai
từ đầu vào có cùng độ dài. 


```python
def la_dang_cau(a,b):
    anh_xa_a_sang_b = {}   #tạo hai bảng ánh xạ 
    anh_xa_b_sang_a = {}
    for i in range(len(a)):  #duyệt từng vị trí trong chuỗi
        ki_tu_a = a[i]      # lấy cặp kí tự ở ví trí tương ứng 
        ki_tu_b = b[i]
        if ki_tu_a in anh_xa_a_sang_b:
            if anh_xa_a_sang_b[ki_tu_a] != ki_tu_b :
                return False
        else:
            anh_xa_a_sang_b[ki_tu_a] = ki_tu_b
        if ki_tu_b in anh_xa_b_sang_a:
            if anh_xa_b_sang_a[ki_tu_b] != ki_tu_a:
                return False
        else :
            anh_xa_b_sang_a[ki_tu_b] = ki_tu_a
    return True 
a = input()
b = input()
print(la_dang_cau(a,b)) 
            

```

Viết hàm divide_with_remainder(dividend, divisor) trả về bộ (q, r) sao cho dividend = divisor*q + r và
0 <= |r| < |divisor|           # divide_with_remainder: chia lấy dư    
                                  (số bị chia , số chia)

```python
dividend = int(input())                               #số bị chia
divisor = int(input())                                # số chia
def divide_with_remainder(dividend,divisor):
    if divisor == 0 :
        raise ValueError("Số chia phải khác 0")       #raise dùng để báo lỗi & dừng chương tình ngay lập tức 
    q = dividend // divisor                           # phép // làm tòn xuống âm vô cực 
    r = dividend % divisor                            # phép chia % mặc định rằng trả về số dư cùng dấu với số chia(divisor)
    return (q,r)                                  # ko cần dùng if để ktra lại 
print(divide_with_remainder(dividend,divisor))
    
 
       
```

```python
# cách ko dùng chia nguyên// và chia dư%
def divide_with_remainder(divedend,divisor):      #q:thương , r :dư
    if divisor == 0 :
        return None                               # báo k có kết quả 
    q = 0 
    while True :
        r = divedend - divisor*q 
        if abs(r) < abs(divisor) :
            return(q,r)                            #result : giữ kết quả và xử lý tiếp 
        if divisor > 0 : #số chia dương tăng q (3,1)
            q += 1 
        else :           # số chia âm giảm q(-4,-2)
            q-=1
result = divide_with_remainder(dividend, divisor)
if result is None :
    print("Error.Số chia phải khác 0")
else :
    print(result)
dividend = int(input())
divisor = int(input())
```

```python
def quadratic_roots_count(a,b,c):
    delta = b**2-4*a*c
    if delta > 0 :
        x1 = (-b + (delta)**0.5)/2*a 
        x2 = (-b + (delta)**0.5)/2*a
        return f'Phương trình có hai nhiệm phân biệt là x1={x1} và x2={x2}'
    elif delta == 0 :
        x = -b/(2*a)
        return f'Phương trình có nhiệm kép là x={x}'
    elif delta < 0 :
        return f"Phương trình vô nhiệm"
a,b,c = map(float,input().split())
print(quadratic_roots_count(a,b,c))
```

W5A28 Viết hàm tính góc ∠ABC (độ) từ tọa độ A(x1,y1), B(x2,y2), C(x3,y3).


```python
import math 
def goc_ABC (A,B,C) :
    x1, y1 = A 
    x2, y2 = B 
    x3, y3 = C 
    BA = (x1-x2, y1-y2)
    BC = (x3-x2, y3-y2)
            # Tích vô hướng 
    vo_huong = BA[0]*BC[0] + BA[1]*BC[1]
            # Tích độ dài 
    len_BA = math.sqrt(BA[0]**2 + BA[1]**2)
    len_BC = math.sqrt(BC[0]**2 + BC[1]**2)
    if len_BA == 0 or len_BC == 0 :
        return None 
            # tính góc(radian->độ)
    cos_theta = vo_huong/(len_BC*len_BA)
    cos_theta = max(-1, min(1, cos_theta)) #giá trị cos trong khoảng -1 đến 1 
    tinh_goc = math.degrees(math.acos(cos_theta))
    return tinh_goc
A = tuple(map(float,input().split()))  # nhập từng điểm Nhập A(x,y): 1 0
B = tuple(map(float,input().split()))        
C = tuple(map(float,input().split()))
print(goc_ABC(A,B,C))
```

W5A30 Viết hàm tính hệ số nhị thức C(n, k).


```python
import math 
def to_hop(n,k):
    if k<0 or k>n :
        return 0 
    return math.factorial(n)//(math.factorial(k)*math.factorial(n-k))
n, k = map(int,input().split())
print(to_hop(n,k))
```

```python
def to_hop(n,k):
    if k<0 or k>n :
        return 0 
    k = min(k,n-k)
    res = 1 
    for i in range(1,k+1):
        res = res*(n-i+1)//i
    return res 
n, k = map(int,input().split())
print(to_hop(n,k))
```

Viết hàm sum_1_to_n(n) trả về tổng các số x từ 1 đến n, nếu x có các chữ số đôi một khác nhau 

```python
def hop_le(x):
    s = str(x) 
    for i in range(len(s)-1):
        if s[i] == s[i+1]:
            return False 
    return True 
def sum_1_to_n(n):
    tong = 0 
    for i in range(1, n+1):
        if hop_le(i):
            tong += i 
    return tong 
print(sum_1_to_n)
```
