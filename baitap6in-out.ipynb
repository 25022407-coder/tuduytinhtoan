W6A1 :Nhập vào một dòng gồm các số nguyên. Hãy loại bỏ phần tử trùng nhưng giữ
nguyên thứ tự xuất hiện đầu tiên, rồi in ra list kết quả.
Yêu cầu in ra đúng dạng list Python, ví dụ [1, 2, 3]


```python
numbers = map(int,input().split())
ket_qua = []
for n in numbers :
    if n not in ket_qua : 
        ket_qua.append(n)
print(ket_qua)
```

W6A2 :Nhập một dòng số nguyên. In ra list tổng tích lũy (phần tử i là tổng từ đầu tới vị
trí i), giữ nguyên thứ tự


```python
n = map(int,input().split())
ket_qua = []
tong = 0 
for i in n :
    tong = tong + i 
    ket_qua.append(tong)
print(ket_qua)
```

W6A4 :Nhập một dòng gồm các cặp key:value cách nhau bởi dấu cách. Tạo
dictionary ánh xạ mỗi key -> list các value theo thứ tự xuất hiện. Nếu key lặp
lại, thêm vào list của key đó 


```python
data = input().split()
d = {}
for items in data: 
    key, value = items.split(":")
    if key not in d : 
        d[key] = []
    d[key].append(value)
print(d)

```

W6A5 :Nhập một list số nguyên, In ra dictionary {'positives': ..., 'negatives': ..., 'zeros':
...} (đếm số >0, <0, =0)


```python
numbers = list(map(int,input().split()))
positives = 0 
negatives = 0 
zeros = 0 
for i in numbers: 
    if i > 0 : 
        positives += 1 
    elif i < 0 : 
        negatives += 1 
    else : 
        zeros += 1 
print(f"positives:{positives},negatives:{negatives},zeros:{zeros}")
```

W6A9 :Viết chương trình nhập vào hai dictionary. Sau đó gộp lại thành một dictionary mới
trong đó nếu một key xuất hiện trong cả hai dict, giá trị sẽ được cộng lại, nếu chỉ có
trong một dict thì giữ nguyên. In ra dictionary mới theo thứ tự từ điển tăng dần của
key. 


```python
dict1 = {int(k):int(v) for k,v in(item.split(":") for item in input().split())}
dict2 = {int(k):int(v) for k,v in(item.split(":") for item in input().split())}
ket_qua =  {k: dict1.get(k,0) + dict2.get(k,0) for k in set(dict1) | set(dict2)}
for k in sorted(ket_qua):
    print(k,ket_qua[k])
```

W6A10 :
Viết chương trình nhập vào một danh sách các số nguyên, và số nguyên dương k. In ra
list gồm các cặp số khác nhau có tổng bằng k. Mỗi cặp là một tuple, hai số trong tuple
có thể giống nhau, thứ tự xuất hiện trong danh sách của số thứ nhất nhỏ hơn thứ tự của
số thứ hai. Các tuple trong list được sắp xếp theo thứ tự tăng dần của phần tử đầu tiên
trong tuple.


```python
ds = list(map(int,input().split()))
k = int(input())
ket_qua = []
for i in range (len(ds)) :
    for j in range (i+1,len(ds)):
        if ds[i] + ds[j] == k :
            ket_qua.append((ds[i],ds[j]))
print(sorted(ket_qua))
```

W6A12 :[Tìm phần tử xuất hiện nhiều nhất]
Nhập vào một list các số nguyên. Hãy tìm số xuất hiện nhiều lần nhất trong list.
Nếu có nhiều số cùng có số lần xuất hiện nhiều nhất, hãy in ra số nhỏ nhất
trong số đó.

```python
ds = list(map(int,input().split()))
dem = {}
for so in ds : 
    dem[so] = dem.get(so, 0) + 1 
max_dem = max(dem.values())
ket_qua = min(so for so in dem if dem[so]== max_dem)
print(ket_qua)
```

W6A13 :[Đảo ngược key và value của dictionary]
Nhập vào một dictionary có value là duy nhất (không trùng nhau). Hãy tạo một
dictionary mới bằng cách đảo ngược key và value của dictionary ban đầu. In
ra dictionary mới. 


```python
n = input().split()
d = {}
for i in range(0, len(n), 2):
    key = n[i]
    value = n[i+1]
    d[key] = value
dao = {}
for k, v in d.items():
    dao[v] = k 
print(dao)
```

W6A15 :[Lọc dictionary theo giá trị]
Nhập vào một dictionary có key là chuỗi và value là số nguyên, và một số
nguyên k ở dòng tiếp theo. Hãy tạo một dictionary mới chỉ chứa các cặp key-value từ dictionary ban đầu mà có value lớn hơn k. In ra dictionary mới. 


```python
d1 = {k:int(v) for k,v in (item.split(":") for item in input().split())}
k = int(input())
d2 = {key: value for key, value in d1.items() if value > k}
print(d2)
```
