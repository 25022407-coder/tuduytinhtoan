W6A9
Viết chương trình nhập vào hai dictionary. Sau đó gộp lại thành một
dictionary mới trong đó nếu một key xuất hiện trong cả hai dict, giá trị sẽ
được cộng lại, nếu chỉ có trong một dict thì giữ nguyên. In ra dictionary mới
theo thứ tự từ điển tăng dần của key. 


```python

def to_dict(lst):
    d = {}
    for item in lst : 
        key, value = item.split(":")
        d[int(key)] = int(value)
    return d 
dict1 = to_dict(input().split())
dict2 = to_dict(input().split())
result = {}
for key in set(dict1) | set(dict2):
    result[key] = dict1.get(key,0) + dict2.get(key,0)
for key in sorted(result):
    print(key,result[key])
               
```

```python
dict1 = {int(k): int(v) for k, v in (item.split(":") for item in input().split())}
dict2 = {int(k): int(v) for k, v in (item.split(":") for item in input().split())}
result = {k: dict1.get(k, 0) + dict2.get(k, 0) for k in set(dict1) | set(dict2)}
for k in sorted(result):
    print(k, result[k])
```

W6A10
Viết chương trình nhập vào một danh sách các số nguyên, và số nguyên
dương k. In ra list gồm các cặp số khác nhau có tổng bằng k. Mỗi cặp là
một tuple, hai số trong tuple có thể giống nhau, thứ tự xuất hiện trong danh
sách của số thứ nhất nhỏ hơn thứ tự của số thứ hai. Các tuple trong list được
sắp xếp theo thứ tự tăng dần của phần tử đầu tiên trong tuple.


```python
so = list(map(int,input().split()))
k = int(input())
result = []
seen = set()                
n = len(so)
for i in range(n):
    for j in range(i+1,n):
        if so[i] + so[j] == k : 
            tupple = (so[i],so[j])
            if tupple not in seen : 
                result.append(tupple)
                seen.add(tupple)       # tránh trùng 
result.sort(key=lambda x: x[0])       # lấy phân tử đầu trong tuple để sắp xếp 
print(result)

```

Nhập vào một list các số nguyên. Hãy tìm số xuất hiện nhiều lần nhất trong
list. Nếu có nhiều số cùng có số lần xuất hiện nhiều nhất, hãy in ra số nhỏ
nhất trong số đó.


```python
so = list(map(int,input().split())) 
d = {}
for n in so : 
    if n in d :
        d[n] += 1 
    else : 
        d[n] = 1 
max_d = max(d.values())   
result = min(n for n in d if d[n]==max_d)
print("Số có lần xuất hiện nhiều nhất là:",result)
```

W6A24
Viết hàm đẩy tất cả các số 0 về đầu (về cuối), các số khác vẫn đúng vị trí trước sau:
0 2 3 4 0 0 0 9 => 0 0 0 2 3 2 4 9

```python
def zeros_end(numbers):
    non_zeros = [x for x in numbers if x!= 0 ]
    zeros = [0] * (len(numbers)-len(non_zeros))
    return non_zeros + zeros
numbers = list(map(int,input().split()))
print(zeros_end(numbers))
```

```python
def zeros_start(numbers):
    non_zero = [x for x in numbers if x!= 0]
    zeros = [0]*(len(numbers) - len(non_zero))
    return zeros + non_zero
print(zeros_start(numbers))
```

W6A26
Tính tổng các phần tử cực đại địa phương trong một danh sách, không tính 2 phần tử đầu và cuối  (lớn hơn hai phần tử đứng sát nó)

```python
def tinh_tong_max(n):
    tong = 0 
    for i in range(1,len(n)-1):
        if n[i] > n[i-1] and n[i] > n[i+1]:
            tong += n[i]
    return tong 
n = list(map(int,input().split()))
print(tinh_tong_max(n))
```

W6A28
Cho một dãy các số nguyên, kiểm tra xem nó là dãy tăng thực sự hay giảm, Hay vô danh(không tăng , không giảm)

```python
def kiem_tra(numbers):
    if numbers == sorted(numbers):
        return 'Dãy tăng'
    elif numbers == sorted(numbers,reverse = True):
        return 'Dãy giảm'
    else : 
        return 'Vô danh'
numbers = list(map(int,input().split()))
print(kiem_tra(numbers))
```
