Bài 2 : một khu đất hình chũ nhật có chiều rộng a , chiều dài b. Người ta xây khu vui chơi hình tròn tại ví trí trung tâm khu đất như hình vẽ . Nhập vào a ,b . Hãy tính phần diện tích còn trống để trồng cây xung quanh . Biết pi=3.14 , lấy kết quả 2 chữ số thập phân 

```python
a = float(input("Nhập chiều rộng a: "))
b = float(input("Nhập chiều rộng b: "))
PI = 3.14
r = a / 2 
S_chunhat = a * b 
S_hinhtron = r * PI * r
s_conlai = S_chunhat - S_hinhtron
print(f"Diện tích còn lại : {s_conlai:.2f}")
```

Bài 5: Nhập vào chữ cái hoa A, in ra chữ cái thường liền trước chữ cái a thường tương
ứng của A. Xét TH khi c = ‘A’ là đặc biệt. 

```python
c = input("Nhập vào chữ cái hoa: ")  
if c== "A" :                                                                       
    print("Đây là trường hợp đặc biệt không có chữ cái liền trước a")
else :
    chu_thuong = c.lower()
    ma_so = ord(chu_thuong)
    ma_so_truoc = ma_so - 1
    ket_qua = chr(ma_so_truoc)
    print(f"""chữ cái thường tương ứng là
           {chu_thuong},chữ liền trước nó là :{ket_qua}""")
#để gán số thứ tự cho chữ cái :hàm ord(); để đổi ngược lại từ số thành chữ :chr()
#phải viết tất cả lệnh của print trên 1 dòng ko đc xuống dòng
#nếu code trên hai dòng thì dùng print(f"""nội dung""")

#Nhập dữ liệu từ bàn phím (ép sang kiểu số thực float)
```

Bài 6. Nhập vào 3 số a, b, c. Kiểm tra xem đây có phải là cạnh của 1 tam giác không,
nếu đúng thì tính diện tích của tam giác đó, in ra với 1 chữ số phần thập phân, nếu nó
không phải 3 cạnh của tam giác thì in ra “Khong phai 3 canh cua tam giac”. 


```python
import math2
a = float(input("Nhập cạnh a: "))
b = float(input("Nhập cạnh b: "))
c = float(input("Nhập cạnh c: "))
#ktra cạnh a,b,c có phải cạnh của tam giác ko ,nếu đúng thì tính diẹn tích / sai thì in "ko pải 3 cạnh tam giác"
if a + b > c and a + c > b and b + c > a :
    p = (a + b + c) / 2 
    # tính diện tích theo côcng thức herong
    s_tamgiac = math.sqrt(p * (p-a) * (p-b) * (p-c))
    # căn bậc hai : math.sqrt
    print(f"Diện tích tam giác{s_tamgiac:.1f}")
    #.1f: tức làm tòn 1 chữ số thập phân
else :
    print("Không phải 3 cạnh của tam giác")
```
