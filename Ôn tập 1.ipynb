1 Yêu cầu. Cước phí Taxi Thông Minh.
Viết một hàm duy nhất smart_taxi_fare(car_type, start_hour, km, wait_minutes) tính và trả về total (tổng tiền phải trả,
kiểu số nguyên, đơn vị VND).
Các quy tắc cần cài đặt:
1. Giá mở cửa (0–0.8 km):
- Xe 4 chỗ: 11,000 VND
- Xe 7 chỗ: 12,000 VND
2. Giá theo km sau 0.8 km:
- Xe 4 chỗ: từ 0.8→20 km ơnh 13,000 VND/km; trên 20 km tính 10,500 VND/km
- Xe 7 chỗ: từ 0.8→20 km ơnh 14,000 VND/km; trên 20 km tính 12,000 VND/km
3. Phí chờ: 5 phút đầu miễn phí; sau đó, mỗi 5 phút (làm tròn lên theo từng block 5 phút) tính 2,000 VND.
4. Phụ thu ban đêm: nếu start_hour thuộc [22,23,0,1,2,3,4,5], cộng +20% trên tiền tạm tính (mở cửa + quãng đường +
chờ).
5. Điều kiện đặc biệt:
- Nếu km > 30 và không phải ban đêm, áp dụng giảm 10% trên tổng tiền.
- Nếu tổng tiền cuối cùng < 50,000 nhưng km > 5, đặt tổng tiền bằng 50,000 VND (mức tối thiểu)

```python
import math
def smart_taxi_fare(car_type, start_hour, km, wait_minutes):
    total = 0 
    if car_type == 4 :
        total += 11000
    elif car_type == 7 :
        total += 12000 
    if km > 0.8 :
        extra_km = km - 0.8 
        if car_type == 4 :
            if km <= 20 :
                total = extra_km *13000
            else :
                total += (20-0.8)*13000
                total += (km-20)*10500
        elif car_type == 7 :
            if km <20 :
                total = extra_km *14000
            else :
                total += (20-0.8)*14000
                total += (km -20)*12000
    if wait_minutes > 5 :
        blocks = math.ceil((wait_minutes-5)/5)
        total += blocks * 2000 
    night_hours = [22,23,0,1,2,3,4]
    is_night = start_hour in night_hours 
    if is_night :
        total *= 1.2 
    if km > 30 and not is_night :
        total *= 0.9 
    if total <50000 and km > 5 :
        total = 50000
    return int(total) 
car_type = int(input())
start_hour = int(input())
km = float(input())
wait_minutes = float(input())
print(smart_taxi_fare(car_type, start_hour, km, wait_minutes))













    
```

2 Yêu cầu: Cho số nguyên x, trả về True nếu tồn tại số nguyên n ≥ 1 sao cho x = 1 + 2 + ... + n, ngược lại trả về False.
Khuyến khích dùng vòng lặp đơn giản để cộng dồn tổng. Tức x = n*(n+1)/2   (số tam giác _ triangular number)

```python
def triangular_number(x):
    if x < 1 :
        return False 
    tong = 0 
    n = 1 
    while tong < x :
        tong += n 
        n += 1 
        if tong == x :
            return True  
    return False 
x = int(input())
print(triangular_number(x))
```

sách được sắp xếp tăng dần.
Quy tắc:
- Nếu a == 0 và b != 0: trả về nghiệm duy nhất [-c / b].
- Nếu a == 0 và b == 0: trả về [] (không có nghiệm hữu hạn duy nhất nếu c == 0, hoặc vô nghiệm nếu c != 0).
- Với a != 0, dùng biệt thức d = bb - 4ac:
- d < 0 → []
- d == 0 → [ -b / (2a) ]
- d > 0 → hai nghiệm [r1, r2] với r1 <= r2

```python
def giai_phuong_trinh(a,b,c):
    if a==0 :
        if b !=0:
            return [-c/b]
        else : 
            return []
    else:
        d = b**2-4*a*c
        if d<0 :
            return []
        elif d == 0 :
            return [-b/(2*a)]
        elif d > 0 :    #danh sách tăng dần
            r1 = (-b + d**0.5)/2*a
            r2 = (-b - d**0.5)/2*a
            return sorted([r1, r2])
a,b,c = map(float,input().split())
print(giai_phuong_trinh(a,b,c))

```

Bài 16 Viết hàm trả về góc ∠ABC theo đơn vị độ khi biết tọa độ các điểm A(x1,y1), B(x2,y2), C(x3,y3)

```python
def tinh_goc(A,B,C):
    x1, y1 = A 
    x2, y2 = B 
    x3, y3 = C 
    BA = (x1-x2,y1-y2)
    BC = (x3-x2,y3-y2)
    tich_vo_huong = BA[0]*BC[0] + BA[1]*BC[1]
    tich_do_dai = (BA[0]**2+BA[1]**2)**0.5 +(BC[0]**2+BC[1]**2)**0.5
    if tich_do_dai == 0 :
        return None 
    else :
        phi = tich_vo_huong / tich_do_dai 
        phi = max(-1,min(1,phi))
        tinh_goc = math.degrees(math.acos(phi))
        return tinh_goc 
A = tuple(map(float,input().split()))
B = tuple(map(float,input().split()))
C = tuple(map(float,input().split()))
print(tinh_goc(A,B,C))
```

và False nếu ngược lại.
- Đầu vào:
- quad: danh sách hoặc bộ gồm đúng 4 đỉnh [(x1,y1), (x2,y2), (x3,y3), (x4,y4)] theo thứ tự (tứ giác có thể lồi hoặc lõm,
nhưng không tự cắt).
- p: một điểm (x, y).
- Đầu ra: giá trị boolean.
- Gợi ý: Cài đặt phép kiểm tra điểm trong đa giác tổng quát (ví dụ: ray casting), đồng thời tính cả trường hợp điểm nằm
trên biên là ở trong

```python
def on_segment(a,b,p):                                  #segment : đoạn 
    (x1, y1), (x2,y2) = a, b 
    x, y = p 
    return (min(x1, x2)) <= x <= max(x1, x2) and min(y1, y2) <= y <= max(y1,y2) and (x2-x1)*(y-y1) == (y2 - y1)*(x-x1)
def point_in_quadrilateral(quad, p):
    x, y = p 
    n = 4 
    inside = False 
    for i in range (n):
        a = quad[i]                                  
        b = quad[(i+i)%n]
        if on_segment(a,b,p): 
            x1,y1 = a 
            x2,y2 = b 
        if on_segment(a,b,p):
            return True 
        x1,y1 = a 
        x2,y2 = b                                         
        if (y1>y) != (y2>y):
            x_intersect = x1 + (y-y1)*(x2-x1) / (y2-y1)  #intersect : giao nhau 
            if x < x_intersect :
                inside = not inside
    return inside 
data = list(map(int,input().split()))
quad = [(data[i], data[i+1]) for i in range(0,8,2)]
p = (x,y)
print(point_in_quadrilateral(quad, p))
                

           
```

- Đầu vào: các số thực a, b, c.
- Đầu ra: 0, 1, hoặc 2. Nếu phương trình suy biến thành 0 = 0 (tức a == 0, b == 0, c == 0), trả về float('inf').

- Nếu a != 0, dùng biệt thức D = b^2 - 4ac.

- Nếu a == 0 và b == 0 và c != 0, phương trình vô nghiệm. 

```python
def quadratic_roots_count(a, b, c) :
    if a == 0 :
        if b!= 0 :
            return "đó là phương trình bậc nhất với đúng một nghiệm thực"
        if b == 0 :
            return " phương trình vô nghiệm."
    else :
        D = b**2 - 4*a*c
        if D < 0 : 
            return 'phương trình vô nhiệm'
        elif D > 0 : 
            return 'Phương trình có hai nhiệm phân biệt'
        else :
            return 'Phương trình có nhiệm kép '
a,b,c = map(int,input().split())
print(quadratic_roots_count(a, b, c))
```

- Đầu vào: chuỗi s.
- Đầu ra: một trong các giá trị {'int', 'float', 'neither'}.
- Quy tắc:
- Cho phép có dấu + hoặc - ở đầu.
- Số nguyên phải chỉ gồm các chữ số và không có dấu chấm thập phân (ví dụ: '100').
- Số thực phải có dấu chấm và có chữ số ở cả hai phía của dấu chấm (ví dụ: '100.0', '-3.14').
- Các chuỗi như '.5', '5.', '1e3' được xem là 'neither' trong bài này.

```python
def detect_number_type(s):                      #.startswith(kí_tự) : Bắt đầu với 
    stemp_s = s 
    if stemp_s.startswith("+") or stemp_s.startswith("-") :
        stemp_s = s[1:]  
    if "." in stemp_s :  #kiểm tra float
        parts = s.split('.')
        if len(parts)==2 and parts[0].isdigit() and parts[1].isdigit():   #hai phía dấu chấm
            return 'float'
        return 'Neither'
    elif stemp_s.isdigit(): #kiểm tra int
        return 'int'
    else :
        return 'neither'
s = input()
print(detect_number_type(s))                                               #parts = s.split('.') : cắt chuỗi bởi dấu chấm  - parts[0]: phần trc dấu chấm
                                                                                            # - parts[1]:phần sau dấu chấm
                                                #[1:] lấy tất cả từ 1 cho đến hết
```

```python
def is_right_triangle(a, b, c) :
    a, b, c = sorted([a,b,c])  #c_max
    sai_so_nho = 1e-6
    return abs(a*a + b*b -c*c) < sai_so_nho 
a, b, c = map(float,input().split())
print(is_right_triangle(a, b, c))
```

- Đầu ra: kết quả số học.
- Yêu cầu thêm:
- Nếu op là '/' và b == 0, phát sinh ValueError.
- Nếu op không hợp lệ, phát sinh ValueError.


```python
def compute(a, b, op) :
    if op == '+' :
        return a + b 
    elif op == '-':
        return a - b 
    elif op == '*':
        return a*b
    elif op == '/':
        if b != 0 :
            return a/b
        else:
            raise ValueError("giá trị b ko hợp lệ")
    else :
        raise ValueError("op không hợp lệ")
try :
    a , b = map(int,input().split())
    op = input()
    print(compute(a, b, op))
except ValueError as e:   #không dùng ngoặc đơn
    print(e)

    

            
```

35 Yêu cầu: Viết hàm divide_with_remainder(dividend, divisor) trả về bộ (q, r) trong đó q là thương nguyên và r là số dư sao
cho dividend = divisor*q + r với 0 <= |r| < |divisor| (dùng ngữ nghĩa của Python với // và %).
- Đầu vào: số nguyên dividend, divisor (divisor khác 0).
- Đầu ra: bộ hai số nguyên (q, r).                                        #devidend : số bị chia
- Ví dụ: divide_with_remainder(5, 3) -> (1, 2).                           # dividend : số chia


```python
def divide_with_remainder(dividend, divisor) :    
    if divisor == 0 :
        raise ValueError ("Lỗi không thể chia cho 0")
    else :
        q = dividend // divisor 
        r = dividend % divisor 
        return (q, r)
try : 
    dividend , divisor = map(int,input().split())
    print(divide_with_remainder(dividend, divisor))
except ValueError as e :
    print(e)
```

```python
def tam_giac(a,b,c):
    p = (a + b + c)/2
    if a+b>c and a+c>b and b+c>a :
        return (p*(p-a)*(p-b)*(p-c))**0.5
    else : 
        raise ValueError ("không thể tạo thành tam giác")
try :
    a, b, c = map(int,input().split())
    print(tam_giac(a,b,c))
except ValueError as e :
    print(e)
    
    

```
