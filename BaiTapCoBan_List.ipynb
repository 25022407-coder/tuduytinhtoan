### Bài 3: 
Tìm giá trị trung bình của các số nguyên dương trong một list.
Nếu không có số dương nào thì trả về "invalid"

```python
def Average_positive_numbers_of_list(numbers):
    tong = 0 
    for x in numbers:
        if x > 0 :
            tong += x 
            return tong/len(numbers)
    return "invalid"
numbers = list(map(int,input().split()))
print(Average_positive_numbers_of_list(numbers))
```

### Bài 4: Kiểm tra một list số nguyên có phải là đối xứng hay không
[1, 2, 3, 5, 3, 2, 1]

```python
def kiem_tra_doi_xung(danh_sach):
    n = len(danh_sach)
    for i in range(n//2):
        if danh_sach[i] != danh_sach[n-1-i]:
            return False
    else :
        return True 
lst = list(map(int,input().split()))
print(kiem_tra_doi_xung(lst))
```

### Bài 5: Đưa tất cả các số bằng 0 trong list về đầu của list. Các số khác không thay đổi thứ tự trước sau
3 4 6 7 1 9 0 0 23 0 0 34 67 89
0 0 0 0 3 4 6 7 1 9 23 34 67 89

```python

def zeros_end(numbers):
    list_zeros = [] 
    khac_zeros = []
    for x in numbers : 
        if x == 0 : 
            list_zeros.append(x)
        elif x!= 0 :
            khac_zeros.append(x)
    ket_qua = list_zeros + khac_zeros
    return ket_qua 
numbers = list(map(int,input().split()))
print(zeros_end(numbers))
```

### Bài 8: Kiểm tra một danh sách số nguyên có phải là ziczac khong
1 3 2 5 2 8 7 9 2 100: ziczac
34 2 56 7 90 20 30 7 120: ziczac
2 2 34 4 56 7 89 : không phải ziczac


```python
def kiem_tra_ziczac(ds):
    tang = True
    giam = True
    for i in range(len(ds)-1):
        if ds[i] > ds[i+1] :
            tang = False 
        if ds[i] < ds[i+1]:
            giam = False 
    if not giam and not tang :
        return "là ziczac" 
    elif giam and not tang or tang and not giam : 
        return "Không phải ziczac"
ds = list(map(int,input().split()))
print(kiem_tra_ziczac(ds))

           
        
```

### Bài 9: Tìm và in ra tất cả các vị trí của số nguyên k trong một danh sách số nguyên cho trước. Nếu không có k trong ds thì in ra "-1".
VD: k = 3, [1,3,4,4,3,3,4,3,6,1] => k xuất hiện tại các vị trí (trả về 1 list) [1, 4, 5, 7]
k = 10, [1,3,4,4,3,3,4,3,6,1] => k không xuất hiện trong ds => trả về None

```python
def find_pos_of_k(lst, k):
    ket_qua = []
    for x in range(len(lst)):
        if lst[x] == k :
            ket_qua.append(x)
    if len(ket_qua) == 0:
        return -1
    else:
        return ket_qua 
lst = list(map(int,input().split()))
k = int(input())
print(find_pos_of_k(lst, k))
```

### Bài 11: 
Cho một danh sách các số nguyên có giá trị trong đoạn [0,9]. Liệt kê số lần xuất hiện của các số có mặt trong danh sách. Số nào xuất hiện trước thì liệt kê trước
VD 1 2 1 2 1 1 1 1

output: 
1: 6
2: 2


```python
chuoi = list(map(int,input().split()))
d = {}
for x in chuoi : 
    d[x] = d.get(x, 0) + 1 
for x in d :
    print(f"{x}:{d[x]}", end=" ")    

```

### Bài 11B: 
Cho một danh sách các số nguyên có giá trị trong đoạn [0,9]. Liệt kê số lần xuất hiện của các số có mặt trong danh sách. Số nào nhỏ thì liệt kê trước
VD 2 1 2 1 2 1 1 1 1

output: dict 



```python
chuoi = list(map(int,input().split()))
d = {}
for x in chuoi :
    d[x] = d.get(x, 0) + 1 
ket_qua = {}
for x in sorted(d.keys()):
    ket_qua[x] = d[x]
print(ket_qua)
```

### Bài 12: 
Cho một chuỗi kí tự chỉ gồm các kí tự chữ cái thường 'a'....'z'. Hãy liệt kê số lần xuất hiện của các kí tự có mặt trong chuỗi theo thứ tự tăng dần.
