Bài 4: Viết chương trình in ra hình tam giác cân có độ cao h được nhập từ bàn phím
Ví dụ: Ta nhập h = 4:
       *
    *     *
 *           *
* * * * * * * * 

```python
h = int(input())
for i in range(1,h+1):
    for j in range(h-i):
        print(" ",end="")
    for k in range(2*i-1):
        print("*", end="")
    print()
```

Bài 16: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu chữ số lẻ 


```python
n = int(input())
chuoi = str(n)
chan = 0 
le = 0 
for so in chuoi :
    if int(so)%2 ==0:
        chan += 1 
    elif so == 0 :
        chan += 1 
    else :
        le += 1 
print(f"số chẵn, số lẻ là: {chan},{le}")
```

Bài 18: Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 3^k hay không

```python
def luy_thua_3(n):
    if n<1 :
        return False 
    while n%3==0 :
        n//=3
    return n==1 
n = int(input())
print(luy_thua_3(n))
```

```python
def luy_thua_3(n):    #dùng lũy thừa tăng dần
    if n<1 :
        return False 
    i = 1
    while i < n :
        i*= 3 
    return i == n 
n = int(input())
print(luy_thua_3(n))
    
```

Bài 19: Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số
kế tiếp sẽ bằng tổng hai số trước đó. Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt quá A

```python
A = int(input())
a, b = 1, 1 
while b < A :
    a,b = b , a+b 
print("số thỏa mãn là :",a)
```

Bài 20: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số là số nguyên tố.

```python
def so_nguyen_to(n):
    chuoi = str(n)
    dem = 0 
    list_nguyen_to = ['2','3','5','7']
    for so in chuoi :
        if so in list_nguyen_to :
            dem += 1 
    return dem
n = int(input())
print(so_nguyen_to(n))
            
```
