Bài 5: Viết chương trình giải phương trình bậc 2 (ax2 + bx + c = 0)

```python
import math 
a = float(input())
b = float(input())
c = float(input())
def giai_phuong_trinh(a,b,c) :
    if a==0 :
       x = -c/b
       return f"Phương tình có nhiệm duy nhât : x = {x}"
    elif a != 0:
        delta = (b**2)-4*a*c
    if delta > 0: 
            x1 = (-b + (delta)**0.5)/(2*a)
            x2 = (-b - (delta)**0.5)/(2*a)
            return f"Phương trình có hai nhiệm phân biệt là : x1={x1} x2={x2}"
    elif delta < 0 :
            return f"Phương trình vô nhiệm"
    elif delta == 0 :
            x3 = -b/(2*a)
            return f"Phương trình có nhiệm kép là: x3={x3}"
print(giai_phuong_trinh(a,b,c))
```

Bài 8: Giải hệ phương trình tuyến tính:
ax  by  m
cx  dy  n

```python
a = float(input())
b = float(input())
c = float(input())
d = float(input())
n = float(input())
m = float(input())
def giai_he_phương_trinh(a , b ,c ,d ,n ,m ):
    dx = m*d - b*n 
    dy = a*n - c*m
    d = a*d-c*b
    if d != 0 :
        x = dx/d
        y = dy/d 
        return f"Hệ phương trình có nhiệm duy nhất là : x ={x} y={y}"
    elif dx == 0 and dy == 0 :
        return f"Phương trình có vô số nhiệm"
    else :
        return f"Phương trình vô nhiệm"
print(giai_he_phương_trinh(a,b,c,d,n,m))
```

Bài 9: Nhập vào thời gian 1 công việc nào đó là x giây. Hãy chuyển đổi và viết ra màn hình số thời gian trên dưới dạng bằng bao nhiêu giờ, bao nhiêu phút, bao nhiêu giây.
